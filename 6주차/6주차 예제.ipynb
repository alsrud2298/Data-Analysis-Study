**구남이**

```python
import seaborn as sns
df = sns.load_dataset('iris')
print(df)
```

```python
# petal_length, petal_width열을 지정하여 선그래프 그리기
df_temp = df[['petal_length','petal_width']]
df_temp.plot()
```

```python
# versicolor와 virginica만 추출하고 x축에 꽃받침 길이(sepal_length), 
# y축에 꽃받침 너비(sepal_width)를 지정하여 산점도 그리기
versi = df[df['species'] == 'versicolor']
virgi = df[df['species'] == 'virginica']
df_temp = pd.concat([versi,virgi])
df_temp.plot(x = 'sepal_length',y = 'sepal_width',kind='scatter')
```

```python
# 전체 데이터에서 petal_length, petal_width 열을 선택하여 박스 플롯 그리기
df[['petal_length','petal_width']].boxplot()
```

```python
# 전체 열의 표준편차 구하기
df.std()
# setosa만 추출해서 전체 열의 평균 구하기
df[df['species'] == 'setosa'].mean()
```

**박민영**

```python
# 1. 다음 csv 파일을 데이터프레임으로 변환하여 변수 df1에 저장하여라.
url =  "https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv"
df1 = pd.read_csv(url)
```

```python
# 2. 데이터프레임 df에서 Manufacturer, Model, Type, Min.Price, Max.Price 열만 선택하여, 5번째 행마다 출력하여 df2에 저장하여라.
# 즉 0, 5, 10, ... 행을 출력하여라.
idx = range(0,len(df1),5)

df2 = df1[['Manufacturer','Model', 'Type', 'Min.Price', 'Max.Price']].iloc[idx]
df2
```

```python
# 3. 데이터프레임 df2의 행 인덱스를 세 개의 변수 Manufacturer, Model, Type의 값으로 생성하라. (단, 세 변수의 값은 _로 구분되어져 있다)
# (ex) Acura_Integra_Small
df2.index = df2['Manufacturer'] + '_' + df2['Model'] + '_' + df2['Type']
df2
```

```python
# 4. df1의 Min.Price와 Max.Price의 산점도를 그려라.
df1[['Min.Price','Max.Price']].plot()
```
