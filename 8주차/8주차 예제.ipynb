```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
```

```python
#1. 두 데이터프레임을 고객번호를 기준으로 병합하여 df3에 저장하시오.
df3 = pd.merge(df1,df2)
df3
```

```python
# 2. df3를 고객 이름을 기준으로 그룹화 한 후 이름이 둘리이고 금액이 20000원 이상인 데이터만 추출하시오.
grouped = df3.groupby(['이름'])
df4 = grouped.get_group('둘리')
df4[df4.금액 >= 20000]
```

```python
# 라이브러리 불러오기
import pandas as pd

# 데이터프레임 만들기
score_df = pd.DataFrame({'name':['Emily','Olivia','chole','Mia'],
                      'mid':[95, 80, 85, 70],
                      'final':[97, 92, 58, 84],
                      'assign':[100, 95, 98, 32]})
score_df
```

```python
# 1. 중간고사, 기말고사, 과제를 각각 30%, 40%, 30%씩 반영하여
# 90점 이상인 경우 A, 80점 이상이면 B, 70점 이상이면 C, 그 이하면 D를 부여하는 
# 함수 'grading' 정의하기
def grading(mid,final,assign):
    score = mid*0.3 + final*0.4 + assign*0.3
    if  score >= 90:
        return 'A'
    elif score >=80:
        return 'B'
    elif score >=70:
        return 'C'
    else:
        return 'D'
```

```python
# 2. score_df에 힉생들의 학점을 계산한 'grade' 열 추가하기 (apply 함수 사용)
score_df['grade'] = score_df.apply(lambda x: grading(x.mid,x.final,x.assign), axis = 1)
score_df
```

```python
# 3. 불린 인덱싱으로 학점이 'A','B'인 학생들만 추출하기
filter = (score_df.grade == 'A')|(score_df.grade == 'B')
score_df.loc[filter,:]
```

```python
# 4. isin() 메소드를 활용하여 학점이 'A','B'인 학생들만 추출하기
score_df[score_df.grade.isin(['A','B'])]
```

```python
# 1. 아래 데이터 프레임을 이용하여 차변잔액, 대변잔액 칼럼을 생성하고자 한다.
# 이 때 차변잔액은 차변합계 - 대변합계이며, 그 값이 음수이면 0으로 처리한다.
# 대변잔액는 대변합계 - 차변합계이며, 그 값이 음수이면 0으로 처리한다.

# 데이터프레임 생성
import pandas as pd
df = pd.DataFrame({'계정코드':['10100', '10300', '10301', '10400', '10600'],
                   '계정과목':['현금', '보통예금', '보통예금', '제예금', '기타단기금융상품'],
                   '빈도':[166, 2169, 2, 35, 1],
                   '차변합계':[367803972, 117268900714, 449982824, 12995303982, 0],
                   '대변합계':[323227674, 116469441606, 0, 10460000000, 14172728]})
df
```

```python
df['차변잔액'] = df.apply(lambda x : x['차변합계'] - x['대변합계']
                            if (x['차변합계'] >= x['대변합계'])
                            else 0, axis = 1 )
df['대변잔액'] = df.apply(lambda x : x['대변합계'] - x['차변합계']
                            if (x['대변합계'] >= x['차변합계'])
                            else 0, axis = 1 )
df
```

```python
# 2. 아래 데이터 프레임을 이용하여 가격이 150원 이하이면 LOW, 250원 이하이면 MEDIUM, 400원 이하이면 HIGH의 값을 가지는 class 변수를 생성하라.

df = pd.DataFrame({'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
                   'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
                   'price': [100, 200, 250, 300, 150, 200, 400],
                   'quantity': [1, 2, 3, 4, 5, 6, 7]})
print(df, '\n')
```

```python
def price(x):
    if  x <= 150:
        return 'LOW'
    elif x <= 250:
        return 'MEDIUM'
    elif x <= 400:
        return 'HIGH'
    
df['class'] = df.price.apply(lambda x : price(x))
df
```

```python
# 3. 2번에서 생성한 데이터프레임을 이용하여 class 별로 총 quantity를 구한다.
df.groupby('class').quantity.sum()
```

```python
# 4. 2번에서 생성한 데이터프레임을 이용하여 city, fruits로 그룹화하여 평균을 구한다.
df.groupby(['city','fruits']).mean()
```

```python
# 5. 2번에서 생성한 데이터프레임을 이용하여 city, fruits로 그룹화한 후, 
# 그룹별로 총 가격(price x quantity)을 계산한다.
df.groupby(['city','fruits']).apply(lambda x: (x.price * x.quantity).sum())
```
