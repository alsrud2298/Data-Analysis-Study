```python
# 2. 판다스의 get_dummies 함수를 이용해 income 열은 Label Encoding하여 y에 저장하고 , 
# 나머지 열은 One-Hot Encoding 하여 X에 저장하시오.
import pandas as pd
from sklearn.preprocessing import LabelEncoder
cols = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
       'occupation', 'relationship', 'race', 'gender', 'capital-gain', 'capital-loss',
       'hours-per-week', 'native-country', 'income']
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
df = pd.read_csv(url, header=None, names=cols, na_values=' ?')
df = df.dropna()
data_cols = ['age','workclass','education','occupation','gender','hours-per-week','income']
adult_df = df[data_cols]
adult_df.head()
```

```python
# income
le = LabelEncoder()
le.fit(adult_df['income'])
y = le.transform(adult_df['income'])

# 나머지 칼럼
X = pd.get_dummies(adult_df[adult_df.columns[:-1]])

# 인코딩 확인
df = X.copy()
df['income'] = y
df.head()
```

```python
# 3. 위에 인코딩한 데이터의 age열과 hours-per-week 열에 대해서 MinMaxScaler를 적용하시오.
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df[['age','hours-per-week']])
age_hour_scaled = scaler.transform(df[['age','hours-per-week']])

age_hour_df_scaled = pd.DataFrame(data=age_hour_scaled, columns=['age','hours-per-week'])
age_hour_df_scaled
```

```python
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

# 와인 데이터를 로드
wine = load_wine()

# wine.data는 wine 데이터 세트에서 피처만으로 된 데이터를 numpy로 가지고 있다. 
wine_data = wine.data

# wine.target은 wine 데이터 세트에서 레이블 데이터를 numpy로 가지고 있다. 
wine_label = wine.target

# Q1. wine label값과 wine label명을 출력해보세요
print('wine label값 : ', wine_label)
print('wine label명 : ', wine.target_names )
```

```python
# Q2. 와인 데이터 세트를 자세히 보기 위해 DataFrame으로 변환해보세요
wine_df = pd.DataFrame(wine_data,columns = wine.feature_names)
wine_df.head()
```

```python
# Q3. 데이터 30% 와 테스트 데이터 세트 70% 로 분리하세요
X_train,X_test,y_train,y_test = train_test_split(wine_data,wine_label,test_size=0.3)
```

```python
# Q4. 의사 결정 트리를 이용해 학습과 예측을 수행해보세요.
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
pred = dt.predict(X_test)
pred
```

```python
# Q5. 예측 결과를 기반으로 의사 결정 트리의 정확도를 측정해보세요.
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

#### 강수민

```python
# 데이터 불러오기
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()

# Q1. cancer_data로 decisiontree분류 분석을 하기 위해 학습/테스트 데이터 세트를 분리해라.
# (test_size=0.3, random_state=11)
# shape로 학습/테스트 데이터 개수 확인
X = cancer_data.data
y = cancer_data.target

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=11)
print(X_train.shape, X_test.shape)
```

```python
# Q2. decisiontree 분류 분석 시행하기(예측 정확도를 print하라)
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
pred = dt.predict(X_test)
accuracy_score(y_test,pred)
```

```python
# Q3. GridSearchCV(빈칸 채우기)
from sklearn.model_selection import GridSearchCV
import pandas as pd

# 파라미터 설정
parameters = {'max_depth':[1, 2, 3], 'min_samples_split':[2, 3]}

# param_grid의 하이퍼 파라미터를 3개의 train, test set fold로 나누어 테스트 수행 설정하기.(refit=True) 
grid = GridSearchCV(dt, param_grid = parameters, cv = 3, refit=True)

# 학습 데이터로 param_grid의 하이퍼 파라미터를 순차적으로 학습하고 평가하기
grid.fit(X_train, y_train)

# GridSearchCV 결과를 추출해 DataFrame으로 변환하기
scores_df = pd.DataFrame(grid.cv_results_)
scores_df[['params', 'mean_test_score', 'rank_test_score', 'split0_test_score', 'split1_test_score']]
```

```python
# Q4. GridSearchCV의 학습된 estimator로 테스트 데이터 세트 정확도 측정하기
estimator = grid.best_estimator_

pred = estimator.predict(X_test)
print(accuracy_score(y_test, pred))
```

#### 박민영

```python
# 물고기 종에 따른 분류를 하기 위한 데이터 로딩
# weight : 무게 / Length : 길이 / Diagonal : 사선 길이 / Height : 높이 / width : 너비

import pandas as pd
import numpy as np

df = pd.read_csv('https://bit.ly/fish_csv')
print(df.head())
```

```python
# Q1. 데이터를 피처 데이터셋 X_fish_df와 타겟 데이터셋 y_fish_df로 나누세요.
X_fish_df = df.iloc[:,1:]
y_fish_df = df['Species']
```

```python
# Q2. 전체 데이터를 학습 데이터셋과 테스트 데이터셋으로 분리하세요.
# 단, test_size = 0.2, random_state = 11로 지정하세요.

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_fish_df,y_fish_df,test_size=0.2,random_state=11)
```

```python
# Q3. 로지스틱 회귀 모델을 이용하기 위해 아래 빈칸을 채워넣으세요.
# 단, 모델 성능 평가 지표로는 정확도를 사용하세요.

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_clf = LogisticRegression(solver = 'liblinear')

lr_clf.fit(X_train,y_train)
lr_pred = lr_clf.predict(X_test)
print(accuracy_score(y_test,lr_pred))
```

```python
# Q4. cross_val_score를 이용하여 StratifiedKFold를 수행하기 위해 아래 빈칸을 채워넣으세요.
# 단, 폴드의 갯수는 5개로 하세요.

from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr_clf, X_train, y_train, scoring='accuracy', cv=5)
for iter_count, accuracy in enumerate(scores):
    print('교차 검증 {0} 정확도 : {1:0.4f}'.format(iter_count, accuracy))
    
print('평균 정확도 {0:0.4f}'.format(np.mean(scores)))
```
